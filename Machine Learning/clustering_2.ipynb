```python
import pandas as pd

nba = pd.read_csv('nba_2013.csv')
```

```python
nba.info()
```

```python
pg_players = nba[nba['pos'] == 'PG']
```

```python
pg_players.head()
```

```python
pg_players.info()
```

```python
pd.options.mode.chained_assignment = None
pg_players['ppg'] = pg_players['pts'] / pg_players['g']
```

```python
pg_players.head()
```

```python
pg_players = pg_players[pg_players['tov'] != 0]
pg_players['atr'] = pg_players['ast'] / pg_players['tov']
```

```python
pg_players.head()
```

```python
pg_players.info()
```

```python
import matplotlib.pyplot as plt

plt.scatter(pg_players['ppg'], pg_players['atr'], c='y')
plt.title('Point Guards')
plt.xlabel('Points Per Game', fontsize=13)
plt.ylabel('Assists Turnover Ratio', fontsize=13)
plt.show()
```

```python
import numpy as np

num_clusters = 5
random_initial_points = np.random.choice(pg_players.index, size=num_clusters)

centroids = pg_players.loc[random_initial_points]
```

```python
plt.scatter(pg_players['ppg'], pg_players['atr'], c='y', label='Point Guards')
plt.scatter(centroids['ppg'], centroids['atr'], c='b', label='Centroids')
plt.title('Point Guards')
plt.xlabel('Points Per Game', fontsize=13)
plt.ylabel('Assists Turnover Ratio', fontsize=13)
plt.legend()
plt.show()
```

```python
centroids_dict = {i: (row.ppg, row.atr) for i, row in enumerate(centroids.itertuples(index=False))}
print(centroids_dict)
```

```python
def euclidean_distance(player, centroid):
    return np.sqrt((player['ppg'] - centroid[0])**2 + (player['atr'] - centroid[1])**2)

# Calculate distances
distances = []
for i, player in pg_players.iterrows():
    player_distances = {}
    min_distance = np.inf
    player_cluster = -1
    for centroid_id, centroid in centroids_dict.items():
        distance = euclidean_distance(player, centroid)
        player_distances[f'centroid_{centroid_id}'] = distance
        if distance < min_distance:
            min_distance = distance
            player_cluster = centroid_id
    distances.append(player_distances)
    pg_players.at[i, 'cluster'] = int(player_cluster)

pg_players['cluster'] = pg_players['cluster'].astype(int)

# Convert distances to DataFrame
distances_df = pd.DataFrame(distances)
print(pg_players)
print(distances_df.head())
```

```python
# Define a color map for the clusters
colors = ['r', 'g', 'b', 'c', 'm']  # Add more colors if needed
pg_players['color'] = pg_players['cluster'].map(lambda x: colors[x])

# Scatter plot the players with color coding according to their cluster
plt.scatter(pg_players['ppg'], pg_players['atr'], c=pg_players['color'])
plt.title('Point Guards')
plt.xlabel('Points Per Game', fontsize=13)
plt.ylabel('Assists Turnover Ratio', fontsize=13)
plt.show()
```

```python
# Update centroids according to the average of each cluster using np.average
new_centroids_dict = {}
for cluster_id in pg_players['cluster'].unique():
    cluster_data = pg_players[pg_players['cluster'] == cluster_id]
    new_centroid = (
        np.average(cluster_data['ppg']),
        np.average(cluster_data['atr'])
    )
    new_centroids_dict[cluster_id] = new_centroid

# Update the original centroids dictionary
centroids_dict.update(new_centroids_dict)

print("Updated Centroids:", centroids_dict)
```

```python
def cluster_and_plot_players(pg_players, centroids_dict):
    def euclidean_distance(player, centroid):
        return np.sqrt((player['ppg'] - centroid[0])**2 + (player['atr'] - centroid[1])**2)

    # Initialize the cluster column
    pg_players['cluster'] = -1

    # Calculate distances and assign clusters
    distances = []
    for i, player in pg_players.iterrows():
        player_distances = {}
        min_distance = np.inf
        player_cluster = -1
        for centroid_id, centroid in centroids_dict.items():
            distance = euclidean_distance(player, centroid)
            player_distances[f'centroid_{centroid_id}'] = distance
            if distance < min_distance:
                min_distance = distance
                player_cluster = centroid_id
        distances.append(player_distances)
        pg_players.at[i, 'cluster'] = int(player_cluster)

    pg_players['cluster'] = pg_players['cluster'].astype(int)

    # Convert distances to DataFrame
    distances_df = pd.DataFrame(distances)
    print(pg_players)
    print(distances_df.head())

    # Define a color map for the clusters
    colors = ['r', 'g', 'b', 'c', 'm']  # Add more colors if needed
    pg_players['color'] = pg_players['cluster'].map(lambda x: colors[x])

    # Scatter plot the players with color coding according to their cluster
    plt.scatter(pg_players['ppg'], pg_players['atr'], c=pg_players['color'])
    plt.title('Point Guards')
    plt.xlabel('Points Per Game', fontsize=13)
    plt.ylabel('Assists Turnover Ratio', fontsize=13)
    plt.show()
```

```python
cluster_and_plot_players(pg_players, centroids_dict)
```

```python
def update_centroids(pg_players, centroids_dict):
    """
    Update centroids according to the average of each cluster using np.average.
    
    Parameters:
    pg_players (pd.DataFrame): DataFrame containing player data with 'ppg', 'atr', and 'cluster' columns.
    centroids_dict (dict): Dictionary containing current centroids.
    
    Returns:
    dict: Updated centroids dictionary.
    """
    new_centroids_dict = {}
    for cluster_id in pg_players['cluster'].unique():
        cluster_data = pg_players[pg_players['cluster'] == cluster_id]
        new_centroid = (
            np.average(cluster_data['ppg']),
            np.average(cluster_data['atr'])
        )
        new_centroids_dict[cluster_id] = new_centroid

    # Update the original centroids dictionary
    centroids_dict.update(new_centroids_dict)

    print("Updated Centroids:", centroids_dict)
    return centroids_dict
```

```python
centroids_dict = update_centroids(pg_players, centroids_dict)
```

```python
cluster_and_plot_players(pg_players, centroids_dict)
```
