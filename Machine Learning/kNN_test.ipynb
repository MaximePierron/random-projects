```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_csv('paris_airbnb.csv')
```

```python
airbnb.info()
```

```python
airbnb.head()
```

```python
new_listing_accomodation = 3
airbnb['distance'] = airbnb[['accommodates']].apply(lambda x: np.abs(x - new_listing_accomodation))
    
```

```python
airbnb.head()
```

```python
airbnb['distance'].value_counts()
```

```python
np.random.seed(1)
```

```python
random_indexes = np.random.permutation(len(airbnb))
print(random_indexes)
```

```python
airbnb = airbnb.iloc[random_indexes]
```

```python
airbnb.head()
```

```python
airbnb = airbnb.sort_values('distance')
```

```python
airbnb['price'][:10]
```

```python
cleaned_price = airbnb['price'].str.replace('$','').replace(',','')
```

```python
cleaned_price
```

```python
airbnb['price'] = pd.to_numeric(cleaned_price, errors='coerce').fillna(0)
```
